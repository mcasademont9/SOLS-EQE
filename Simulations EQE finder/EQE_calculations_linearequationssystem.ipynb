```python
#Here we do the calculations with the linear equations system

index_1 = 1
index_2 = 0
Active_layer_1 = Active_layers[index_1]
Jsc_divWL = np.array(Material_pairs[index_1][index_2].Jsc_Dividing_wavelength_graph_1)
wl_list = np.array(Material_pairs[index_1][index_2].Dividing_wavelength_list)
Solar_data_nparray = np.array(Solar_Data)

spectrum_matrix = np.zeros((len(wl_list), len(wl_list)))
Solar_Data_first_index = np.where(Solar_data_nparray[:,0] == wl_list[0])[0]
Solar_Data_last_index = np.where(Solar_data_nparray[:,0] == wl_list[-1])[0]
wl_interval = wl_list[1]-wl_list[0]

fig, ax = plt.subplots(figsize=(15,7))

Solar_data_wavelength_nparray = np.zeros(Solar_data_nparray.shape)
for j in range(len(Solar_data_nparray[:,0])):
    Solar_data_wavelength_nparray[j,0] = Solar_data_nparray[j,0]
    Solar_data_wavelength_nparray[j,1] = Solar_data_nparray[j,1]/1240*Solar_data_nparray[j,0]

for i in range(len(wl_list)):    
    solar_data_actual_i = np.where(Solar_data_nparray[:,0] == wl_list[i])[0]
    spectrum_matrix[i,0:i] = np.transpose(0.1*wl_interval*Solar_data_wavelength_nparray[Solar_Data_first_index[0]:solar_data_actual_i[0],1])
    

EQE_linearsolution, residuals, rank, s = np.linalg.lstsq(spectrum_matrix, Jsc_divWL, rcond=None)

#print(EQE_linearsolution)
ax.plot(sf.Extract_Column(Active_layer_1.EQE_curve, 0), 1*np.array(sf.Extract_Column(Active_layer_1.EQE_curve, 1)), label = 'Real EQE')
ax.plot(wl_list[:], 10*1/Active_layer_1.Integrated_Jsc_correction_factor*EQE_linearsolution, label = 'Calculated EQE')
ax.legend()
plt.ylim(0,50)
plt.figsize=(8, 6)

```
